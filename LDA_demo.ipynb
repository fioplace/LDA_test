```python
import nltk
import re

stop_list = nltk.corpus.stopwords.words('english')
stemmer = nltk.stem.porter.PorterStemmer()

import gensim

def load_corpus(dir):
    # dir is a directory with plain text files to load.
    corpus = nltk.corpus.PlaintextCorpusReader(dir, '.+\.json')
    return corpus

corpus = load_corpus(path)
```

```python
doc_list = corpus.fileids()
doc_list
```

```python
# convert json to dataframe

import pandas as pd
import ast
import json

def corpus2df(corpus):
    rows = []
    fids = corpus.fileids()
    for fid in fids:
        doc_raw = corpus.raw(fid)
        text = json.loads(doc_raw)
        rows.append(text)
    return rows
        
        

rows = corpus2df(corpus)       
df = pd.DataFrame(rows)
df
```

```python
read_dict = lambda input: pd.Series(input)
tmp = df['sitting_metadata'].apply(read_dict)
df = df.drop('sitting_metadata',axis = 1)
df = df.join(tmp)
df
```

```python
def preprocessing(doc):
    doc = nltk.word_tokenize(doc)
    doc = [w.lower() for w in doc]
    doc = [w for w in doc if re.search('^[a-z]+$', w)]
    doc = [w for w in doc if w not in stop_list]
    doc = [stemmer.stem(w) for w in doc] 
    
    return doc
```

```python
df['processed'] = df['speech'].apply(preprocessing)
```

```python
#pick up speech

#can do filtering before this step

# groups = df.groupby('sitting_date')
# for date, grouped_dataframe in groups:
#     ...


speechs = df['processed'].values
speechs = speechs.tolist()
speechs[0]
```

```python
def docs2vecs(docs, dictionary):
    # docs is a list of documents returned by corpus2docs.
    # dictionary is a gensim.corpora.Dictionary object.
    vecs1 = [dictionary.doc2bow(doc) for doc in docs]
    tfidf = gensim.models.TfidfModel(vecs1)
    vecs2 = [tfidf[vec] for vec in vecs1]
    return vecs2
```

```python
import gensim
dictionary = gensim.corpora.Dictionary(speechs)
doc_vecs = docs2vecs(speechs, dictionary)
```

```python
num_topics = 10
```

```python
doc_lda = gensim.models.ldamodel.LdaModel(corpus=doc_vecs, id2word=dictionary, num_topics=num_topics)
```

```python
topics = doc_lda.show_topics(10, 15)

for i in range(0, 10):
    print(topics[i])
```
